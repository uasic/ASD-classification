```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')

def merge_dataframes(df1, df2):
    import pandas as pd
    # Concatenate the dataframes along the index
    merged_df = pd.concat([df1, df2], axis=0)
    # Group by the index and aggregate the values
    merged_df = merged_df.groupby(merged_df.index).first()
    return merged_df
```

```python
metadata = pd.read_excel("metadata/full_overview_WGS.xlsx", sheet_name="metadata2")[["run", "locationContinent"]].set_index("run")
metadata.dropna(inplace=True)
metadata.head()
```

```python
continents = metadata["locationContinent"].value_counts()
print(continents)
```

```python
inputs = "datasets"
for dirpath, dirnames, filenames in os.walk(inputs):
    for file in filenames:
        if file.endswith(".csv") and "age" not in file and "sex" not in file and "location" not in file:
            file_path = os.path.join(dirpath, file)
            print(f"\nFile: {file}")
            dataframe = pd.read_csv(file_path, low_memory=False).set_index("run")
            #display(dataframe.head(5))
            #print(dataframe.shape)
            metadata_filtered = metadata.loc[metadata.index.intersection(dataframe.index)]
            dataframe = merge_dataframes(metadata_filtered, dataframe)
            #display(dataframe.head(5))
            #print(dataframe.shape)
            
            file = file[:-4]
            
            dataframe_na = dataframe[dataframe['locationContinent'] == "North America"]
            dataframe_na.reset_index(inplace=True)
            rows, columns = dataframe_na.shape
            if rows > 0:
                dataframe_na.to_csv(f"datasets/{file}-location-northAmerica.csv", index=False)
                print("new file:", f"{file}-location-northAmerica.csv", dataframe_na.shape)
            else: print (f"{file}-location-northAmerica.csv doesn't have any samples.")
            del dataframe_na
            
            dataframe_sa = dataframe[dataframe['locationContinent'] == "South America"]
            dataframe_sa.reset_index(inplace=True)
            rows, columns = dataframe_sa.shape
            if rows > 0:
                dataframe_sa.to_csv(f"datasets/{file}-location-southAmerica.csv", index=False)
                print("new file:", f"{file}-location-southAmerica.csv", dataframe_sa.shape)
            else: print (f"{file}-location-southAmerica.csv doesn't have any samples.")
            del dataframe_sa
            
            dataframe_asia = dataframe[dataframe['locationContinent'] == "Asia"]
            dataframe_asia.reset_index(inplace=True)
            rows, columns = dataframe_asia.shape
            if rows > 0:
                dataframe_asia.to_csv(f"datasets/{file}-location-Asia.csv", index=False)
                print("new file:", f"{file}-location-Asia.csv", dataframe_asia.shape)
            else: print (f"{file}-location-Asia.csv doesn't have any samples.")
            del dataframe_asia
            
            dataframe_europe = dataframe[dataframe['locationContinent'] == "Europe"]
            dataframe_europe.reset_index(inplace=True)
            rows, columns = dataframe_europe.shape
            if rows > 0:
                dataframe_europe.to_csv(f"datasets/{file}-location-Europe.csv", index=False)
                print("new file:", f"{file}-location-Europe.csv", dataframe_europe.shape)
            else: print (f"{file}-location-Europe.csv doesn't have any samples.")
            del dataframe_europe
```
