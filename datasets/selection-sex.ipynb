```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')

def merge_dataframes(df1, df2):
    import pandas as pd
    # Concatenate the dataframes along the index
    merged_df = pd.concat([df1, df2], axis=0)
    # Group by the index and aggregate the values
    merged_df = merged_df.groupby(merged_df.index).first()
    return merged_df
```

```python
metadata = pd.read_excel("metadata/full_overview_WGS.xlsx", sheet_name="metadata2")[["run", "sex"]].set_index("run")
metadata.dropna(inplace=True)
metadata.head()
```

```python
sex = metadata["sex"].value_counts()
print(sex)
```

```python
inputs = "datasets"
for dirpath, dirnames, filenames in os.walk(inputs):
    for file in filenames:
        if file.endswith(".csv") and "age" not in file and "sex" not in file:
            file_path = os.path.join(dirpath, file)
            print(f"\nFile: {file}")
            dataframe = pd.read_csv(file_path, low_memory=False).set_index("run")
            #display(dataframe.head(5))
            #print(dataframe.shape)
            metadata_filtered = metadata.loc[metadata.index.intersection(dataframe.index)]
            print(dataframe.shape)
            dataframe = merge_dataframes(metadata_filtered, dataframe)
            #display(dataframe.head(5))
            #print(dataframe.shape)
            
            file = file[:-4]
            
            dataframe_m = dataframe[dataframe['sex'] == "M"]
            dataframe_m.reset_index(inplace=True)
            rows, columns = dataframe_m.shape
            if rows > 0:
                dataframe_m.to_csv(f"datasets/{file}-sex-male.csv", index=False)
                print("new file:", f"{file}-sex-male.csv", dataframe_m.shape)
            else: print (f"{file}-sex-male.csv.csv doesn't have any samples.")
            del dataframe_m
            
            dataframe_f = dataframe[dataframe['sex'] == "F"]
            dataframe_f.reset_index(inplace=True)
            rows, columns = dataframe_f.shape
            if rows > 0:
                dataframe_f.to_csv(f"datasets/{file}-sex-female.csv", index=False)
                print("new file:", f"{file}-sex-female.csv", dataframe_f.shape)
            else: print (f"{file}-sex-female.csv.csv doesn't have any samples.")
            del dataframe_f
            
            del dataframe
```
