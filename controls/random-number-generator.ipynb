```python
import pandas as pd
from random import seed
from random import random
import numpy as np
```

```python
data = pd.DataFrame(index=range(500), columns=range(50))
data.head()
```

```python
data_random = data.applymap(lambda x: np.random.rand() if pd.isna(x) else x)
```

```python
data_random.head(10)
```

```python
# Generate a random list of 250 'TD' and 250 'ASD'
diagnosis_labels = np.random.choice(['TD', 'ASD'], size=500, replace=True)
# Add the 'diagnosis' column to the DataFrame
data_random['diagnosis'] = diagnosis_labels
```

```python
data_random.head(10)
```

```python
data_random.to_csv('random_data.csv')
```
